```python
import json
from matplotlib import pyplot as plt
```

```python
fc = []
with open('self-critical.pytorch/log_fc.txt') as f:
    for line in f:
        fc.append(json.loads(line))
```

```python
fc_bert = []
with open('self-critical.pytorch/log_fc_bert.txt') as f:
    for line in f:
        fc_bert.append(json.loads(line))
```

```python
fc_bert2 = []
with open('self-critical.pytorch/log_fc_bert2.txt') as f:
    for line in f:
        fc_bert2.append(json.loads(line))
```

```python
fc_bert5 = []
with open('self-critical.pytorch/log_fc_bert5.txt') as f:
    for line in f:
        fc_bert5.append(json.loads(line))
```

```python
fc_synonyms2 = []
with open('self-critical.pytorch/log_fc_synonyms2.txt') as f:
    for line in f:
        fc_synonyms2.append(json.loads(line))
```

```python
fc_bert2_p01 = []
with open('self-critical.pytorch/log.txt') as f:
    for line in f:
        fc_bert2_p01.append(json.loads(line))
```

```python
fc_bert2_p06 = []
with open('self-critical.pytorch/log.txt') as f:
    for line in f:
        fc_bert2_p06.append(json.loads(line))
```

```python
spices_fc = [item['SPICE'] for item in fc][:19]
spices_fc_bert = [item['SPICE'] for item in fc_bert][:19]
spices_fc_bert2 = [item['SPICE'] for item in fc_bert2][:19]
spices_fc_bert5 = [item['SPICE'] for item in fc_bert5][:19]
spices_fc_bert2_p01 = [item['SPICE'] for item in fc_bert2_p01][:19]
spices_fc_bert2_p06 = [item['SPICE'] for item in fc_bert2_p06][:19]
spices_fc_synonyms2 = [item['SPICE'] for item in fc_synonyms2][:19]
```

```python
print (len(spices_fc))
print (len(spices_fc_bert))
print (len(spices_fc_bert2))
print (len(spices_fc_bert2_p01))
```

```python
#plt.plot(range(len(spices_fc)), spices_fc, label='aaa')
#plt.plot(range(len(spices_fc)), spices_fc_bert, label='bbb')
plt.plot(range(len(spices_fc)), spices_fc_bert2, label='03')
plt.plot(range(len(spices_fc)), spices_fc_bert2_p01, label='01')
#plt.plot(range(len(spices_fc)), spices_fc_synonyms2, label='eee')
#plt.plot(range(len(spices_fc)), spices_fc_bert2_p06, label='06')
#plt.plot(range(len(spices_fc)), spices_fc_bert5, label='ddd')
plt.legend()
plt.show()
```

```python
plt.plot(range(len(spices_fc_bert2)), spices_fc)
plt.plot(range(len(spices_fc_bert5)), spices_fc_bert)
plt.show()
```

```python
ciders_fc = [item['CIDEr'] for item in fc][:19]
ciders_fc_bert = [item['CIDEr'] for item in fc_bert][:19]
ciders_fc_bert2 = [item['CIDEr'] for item in fc_bert2][:19]
ciders_fc_bert5 = [item['CIDEr'] for item in fc_bert5][:19]
```

```python
plt.figure(figsize=(7, 7))
plt.plot(range(1, len(spices_fc) + 1), ciders_fc, label='No augmentation')
plt.plot(range(1, len(spices_fc) + 1), ciders_fc_bert, label='Bert augmentation')
plt.plot(range(1, len(spices_fc) + 1), ciders_fc_bert2, label='Bert augmentation x2')
plt.plot(range(1, len(spices_fc) + 1), ciders_fc_bert5, label='Bert augmentation x5')
plt.legend()
plt.title('CIDEr score per epoch')
plt.xticks(range(1, 20))
plt.xlabel('Epoch')
plt.ylabel('CIDEr')
plt.xlim([1, 19])
plt.show()
```
